```python
import cvxpy as cp
import numpy as np

# Generate data.
m = 20
n = 15
np.random.seed(1)
A = np.random.randn(m, n)
b = np.random.randn(m)

# Define and solve the CVXPY problem.
x = cp.Variable(n)
cost = cp.sum_squares(A @ x - b)
prob = cp.Problem(cp.Minimize(cost))
prob.solve()

# Print result.
print("\nThe optimal value is", prob.value)
print("The optimal x is")
print(x.value)
print("The norm of the residual is ", cp.norm(A @ x - b, p=2).value)
```

```python
# 例1.3 计算1+2+……+7并输出
sum = 0
number = int(input("请输入整数："))
print("从小到大排列输出数字：")
for i in range(1, number+1):
    sum += i # 设置sum为i的和 
    print("%d"%(i), end='')
    if i<number:
        print("+", end='') #设置输出连加的算式
    else:
        print("=", end='')
print("%d"%(sum))

sum = 0
print("从大到小排列输出数字：")
for i in range(number, 0, -1):
    sum += i # 设置sum为i的和 
    print("%d"%(i), end='')
    if i>1:
        print("+", end='') #设置输出连加的算式
    else:
        print("=", end='')
print("%d"%(sum))
```

```python
# 例1.4求sin x=x-x^3/3!+x^5/5!+……，直到最后一项的绝对值小于10^-6时停止计算。其中x以弧度为单位，但键盘输入时以角度为单位
# 分析：累计求和问题，关键是如何求累加项，用递推的办法来求累加项
from math import * # 加载数学模块math的所有对象
n = 0
x1 = float(input("请输入角度："))
x = radians(x1)
s = a = x
while abs(a) >= 1e-6:
    a *= -x*x/(2*n+3)/(2*n+2)
    n += 1
    s += a
print("x={}, sin(x)={}".format(x1, s))
```

```python
# 例1.5 先定义求阶乘n!的函数，再调用求5！
def factorial(n):
    r = 1
    while n > 1:
        r *= n
        n -= 1
    return r

print(factorial(5)) # 调用函数
```

```python
# 例1.8 输入n，求n!的值
n = int(input("请输入n的值："))
def fac(n):
    if n <= 1:
        return 1
    else:
        return n*fac(n-1)
m = fac(n) # 调用函数
print("%d!=%5d"%(n,m))
```

```python
# 例1.9 用递归方法计算下列多项式函数的值P(x, n)=x-x^2+x^3+……+(-1)^(n-1)x^n (n>0)
# 分析：函数定义不是递归定义形式，对原来的定义进行如下数学变换
# p(x)=x[1-(x-x^2+x^3-……+(-1)^(n-2)x^(n-1))]=x[1-p(x, n-1)]
# 变换后，将原来的非递归转化为等价的递归定义

x, n = eval(input("请输入x和n的值："))
def p(x,n):
    if n == 1:
        return x
    else:
        return x*(1-p(x, n-1))

v = p(x, n) # 调用函数
print("p(%d, %d)=%d"%(x,n,v))
```

```python
# 例1.9 求y=e^2 + sum(1+ln n+ sin n )/2pi()n=100
# 分析：定义一个匿名函数求累加项，循环控制累加100次
from math import log, exp, sin, pi
f = lambda n:(1 + log(n) + sin(n))/(2 * pi)
y = exp(2)
for n in range(1, 101):
    y += f(n)
print("y=%7.4f"%y)
```

```python
# ufunc函数——通用函数，能对数组中元素逐个进行操作的函数
# 例2.16 ufunc函数效率示例
import numpy as np, time, math
x = [i*0.01 for i in range(1000000)]
start = time.time() # 1970纪元后经过的浮点秒数
for (i,t) in enumerate(x):
    x[i] = math.sin(t)
print("math.sin:", time.time()-start)

y = np.array([i*0.01 for i in range(1000000)])
start = time.time()
y = np.sin(y)
print("numpy.sin:", time.time()-start)
# 对数组的操作，numpy函数整体花费时间少于math模块函数
```

```python
# 例2.47 把四个不同正态分布的密度函数画在4个子窗口
import matplotlib.pyplot as plt 
import numpy as np
from scipy.stats import norm
import os
mu0 = [-1, 0]
s0 = [0.5, 1]
x = np.linspace(-7, 7, 100)
plt.rc('font', size=16)
plt.rc('text', usetex=True)
plt.rc('axes', unicode_minus=False)
f, ax = plt.subplots(len(mu0), len(s0), sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        mu = mu0[i]
        s = s0[j]
        y= norm(mu, s).pdf(x)
        ax[i,j].plot(x,y)
        ax[i,j].plot(1,0,label="$\\mu$ = {:3.2f}\n$\\sigma$ = {:3.2f}".format(mu,s))
        ax[i,j].legend(fontsize=12)
ax[1,1].set_xlabel('$x$')
ax[0,0].set_ylabel('pdf($x$)')
plt.savefig('D:\\学习\\科研\\2019年 生态中心\\数学建模\\python\\Python数学实验与建模-程序及数据\\程序及数据\\02第2章  数据处理与可视化\\figure2_47.png')
plt.show()
```

```python
# 例5.6 已知某种商品6个仓库的存货量，8个客户对该商品的需求量，单位商品运价如表，确定6个仓库到8个客户的商品调运数量，使总的运输费用最小
import cvxpy as cp
import numpy as np
import  pandas as pd
import os
d1 = pd.read_excel("D:\\学习\\科研\\2019年 生态中心\\数学建模\\python\\Python数学实验与建模-程序及数据\\程序及数据\\05第5章  线性规划\\Pdata5_6.xlsx", header=None)
d2 = d1.values
c = d2[:-1,:-1]
d = d2[-1,:-1].reshape(1,-1)
e = d2[:-1,-1].reshape(-1,1)
x = cp.Variable((6,8))
obj = cp.Minimize(cp.sum(cp.multiply(c,x))) # 构造目标函数
con = [cp.sum(x,axis=1,keepdims=True)<=e, cp.sum(x,axis=0,keepdims=True)==d, x>=0] #构造约束条件
prob = cp.Problem(obj,con) # 构造模型
prob.solve(solver='GLPK_MI', verbose=True) # 求解模型
print("最优值为：",prob.value)
print("最优解为：\n",x.value)
```

```python
# 例10.2 如图所示无向图
import numpy as np
import networkx as nx
import pylab as plt
import os
a = np.zeros((5,5))
a[0, 1:5] = [9,2,4,7]
a[1, 2:4] = [3,4]
a[2, [3,4]] = [8,4] # 输入邻接矩阵的上三角元素
a[3, 4] = 6
print(a)
np.savetxt("D:\\学习\\科研\\2019年 生态中心\\数学建模\\python\\Python数学实验与建模-程序及数据\\程序及数据\\10第10章  图论模型\\Pdata10_2.txt", a)
# 保存邻接矩阵供以后使用
i, j = np.nonzero(a) # 提取顶点编号
w = a[i,j] # 提出a中的非零元素
edges = list(zip(i,j,w))
G = nx.Graph()
G.add_weighted_edges_from(edges)
key = range(5)
s = [str(i+1) for i in range(5)]
s = dict(zip(key,s)) # 构造用于顶点标注的字符字典
plt.rc('font', size=18)
plt.subplot(121)
nx.draw(G, font_weight = 'bold', labels=s)
plt.subplot(122)
pos = nx.shell_layout(G) # 布局设置
nx.draw_networkx(G, pos, node_size=260, labels=s)
w = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, font_size=12, edge_labels=w) # 标注权重
plt.savefig("D:\\学习\\科研\\2019年 生态中心\\数学建模\\python\\Python数学实验与建模-程序及数据\\程序及数据\\10第10章  图论模型\\figure10_2.png", dpi=500)
plt.show()
```

```python
#程序文件Pex11_7.py
import numpy as np
from sklearn.decomposition import PCA
import os
a=np.loadtxt("D:\\学习\\科研\\2019年 生态中心\\数学建模\\python\\Python数学实验与建模-程序及数据\\程序及数据\\11第11章  多元分析\\Pdata11_7.txt")
b=np.r_[a[:,1:4],a[:,-3:]]  #构造数据矩阵
md=PCA().fit(b)  #构造并训练模型
print("特征值为：",md.explained_variance_)
print("各主成分的贡献率：",md.explained_variance_ratio_)
print("奇异值为：",md.singular_values_)
print("各主成分的系数：\n",md.components_)  #每行是一个主成分
"""下面直接计算特征值和特征向量，和库函数进行对比"""
cf=np.cov(b.T)  #计算协方差阵
c,d=np.linalg.eig(cf) #求特征值和特征向量
print("特征值为：",c)
print("特征向量为：\n",d)
print("各主成分的贡献率为：",c/np.sum(c))
```
