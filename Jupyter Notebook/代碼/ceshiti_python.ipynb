```python
# 1、作出一元函数图形
import matplotlib.pyplot as plt 
import numpy as np
x = np.linspace(-3,3,1000)
interval0 = [1 if(i>0) else 0 for i in x]
interval1 = [1 if(i<=0) else 0 for i in x]
y = x* interval0 + np.sin(x)*interval1
plt.plot(x,y)
plt.show()
```

```python
# 2、做出二元函数图性
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt 
from matplotlib import cm 
import numpy as np

ax = plt.axes(projection='3d')
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)

X,Y = np.meshgrid(x,y)
Z = np.ones((len(x),len(y)))
for i in range(len(x)):
    for j in range(len(y)):
        # print(Y[j])
        if x[i]**2 + y[j]**2 <= 1:
            Z[i,j] = x[i]**2 + y[j]**2
        else:
            Z[i,j] = np.exp(1 - x[i]**2 - y[j]**2) * np.cos(2 * np.pi * (x[i]**2 + y[j]**2))
            
ax.plot_surface(X,Y,Z)
plt.show()
```

```python
# 2、做出二元函数图性
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt 
import numpy as np

x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
X,Y = np.meshgrid(x,y)
Z = np.ones((len(x),len(y)))
for i in range(len(x)):
    for j in range(len(y)):
        # print(Y[j])
        if x[i]**2 + y[j]**2 <= 1:
            Z[i,j] = x[i]**2 + y[j]**2
        else:
            Z[i,j] = np.exp(1 - x[i]**2 - y[j]**2) * np.cos(2 * np.pi * (x[i]**2 + y[j]**2))

ax1 = plt.subplot(1,2,1,projection='3d')
ax1.plot_surface(X,Y,Z, cmap='viridis')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')
ax2 = plt.subplot(1,2,2, projection='3d')
ax2.plot_wireframe(X,Y,Z,color='c')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('z')      
plt.show()
```

```python
# 作业指派问题
import cvxpy as cp
import numpy as np
c = np.array([[15,18,21,24],[19,23,22,18],[26,17,16,19],[19,21,23,17]])
x = cp.Variable((4,4),integer=True)
obj = cp.Minimize(cp.sum(cp.multiply(c,x)))
con = [0<=x,x<=1,cp.sum(x,axis=0,keepdims=True)==1,cp.sum(x,axis=1,keepdims=True)==1]
prob = cp.Problem(obj,con)
prob.solve(solve='GLPK_MI')
print("最优值为：",prob.value)
print("最优解为：\n",x.value)
```

```python
# 作业最短路径
import numpy as np
import networkx as nx
List = [(1,2,2),(1,3,4),(2,4,3),(2,5,3),(2,6,1),(3,4,2),(3,5,3),(3,6,1),(4,7,1),(5,7,3),(6,7,4)]
G = nx.Graph()
G.add_weighted_edges_from(List)
p = nx.dijkstra_path(G,source=1,target=7,weight='weight')
d = nx.dijkstra_path_length(G,1,7,weight='weight')
print("最短路径为：",p,"；最短距离为：",d)
```

```python
# 6、非线性规划
import cvxpy as cp
import numpy as np
c1 = np.array([-4,0])
c2 = np.array([3,6])
a = np.array([1,1],[2,1],[1,-4])
b = np.array([4,5,-2])
x = cp.Variable(2)
obj = cp.Minimize(c1 * x**2 + c2 * x)
con = [0<=x, a*x<=b]
prob = cp.Problem(obj,con)
prob.solve()
print("最优值为：",prob.value)
print("最优解为：\n",x.value)
```
